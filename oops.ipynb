```python
class student:
    name ="abc"
    anage = 17
    def __init__(self,name,age):
        self.name=name
        self.age=age
    @classmethod
    def get_name(cls):
        print(cls.name)
    @staticmethod
    def wel_to():
        print("hi how are you")
stu1=student("ashu", 18)
stu1.get_name()
stu1.wel_to()
```

```python
class vechicle:
    def __init__(self,gear,color):
        self.gear=gear
        self.color=color
class car(vechicle):
    def __init__(self,gear,color,windows):
        super().__init__(gear,color)
        self.windows=windows
        
c=car(4, "red",8)
print(c.color)
print(c.gear)    
```

```python
class vechicle:
    def __init__(self,gear,color):
        self.gear=gear
        self.color=color
class driver:        
    def __init__(self,name,age):
        self.name=name
        self.age=age   
class car(vechicle,driver):
    def __init__(self,gear,color,windows,name,age):
        super().__init__(gear,color)
        driver.__init__(self,name,age)
        self.windows=windows
        
c=car(4, "red",8,"ashu",17)
print(c.color)
print(c.gear)    
```

```python
from abc import ABC, abstractmethod
class A(ABC):
    @abstractmethod
    def fun1(self):
        pass
class B(A):
    def fun1(self):
        print("hello")
class C(A):
    def fun1(self):
        print("hellooo")
c=B()
c.fun1()
d=C()
d.fun1()
```

```python
#function overloading
class vehicle:
    def grt_detail(self):
        print("this")
class car(vehicle):
    def grt_detail(self):
        print("that")
c=car()
c.grt_detail()
```

```python
#duck typing
class vehicle():
    def grt_detail(self):
        print("this")
class car():
    def grt_detail(self):
        print("that")
c=car()
c.grt_detail()
d=vehicle()
d.grt_detail()
```

```python
while True:
    try:
        num=int(input())
        denom=int(input())
        value=num/denom
        print(value)
        break
    except (ValueError,ZeroDivisionError):
        print("denom should be integer and it shold not be zero")
        
    except:
        print("all good")
```

```python
while True:
    try:
        num = int(input())
        denom = int(input())
        value = num / denom
    except (ValueError, ZeroDivisionError):
        print(" Denominator should be integer and it should not be zero")
    except:
        print(" Something went wrong")
    else:
        # Runs only if no exception occurs
        print("Result:", value)
        break
    finally:
        print("Done")
```
