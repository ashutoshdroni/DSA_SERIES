## looping in list

```python
li = [1,2,"parikh",9,10,11,"ankush"]
li
```

```python
for i in range(len(li)):
    print(li[i])
    
```

```python
for i in range(2,len(li)):
    print(li[i])
```

```python
for ele in li:
    print(ele)
```

```python
for ele in li[2:]:
    print(ele)
```

```python
for ele in li[2:4]:
    print(ele)
```
