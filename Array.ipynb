```python
def largest_arr(arr):
    largest = arr[0]
    for i in range(1,len(arr)):
        if arr[i]>largest:
            largest=arr[i]
    return largest
arr=[1,2,3,45]
largest_arr(arr)
```

```python
def seclargest_arr(arr):
    largest = arr[0]
    seclargest=arr[-1]
    for i in range(1,len(arr)):
        if arr[i]>largest:
            seclargest=largest
            largest=arr[i]
        else:
            if arr[i] != largest and arr[i]> seclargest:
                seclargest=arr[i]
    return seclargest
arr=[1,2,3,45]
seclargest_arr(arr)
```

```python
def secsmallest_arr(arr):
    smallest = arr[0]
    secsmallest=float('inf')
    for i in range(1,len(arr)):
        if arr[i]<smallest:
            secsmallest=smallest
            smallest=arr[i]
        else:
            if arr[i] != smallest and arr[i]< secsmallest:
                secsmallest=arr[i]
    return secsmallest
arr=[1,2,3,45]
secsmallest_arr(arr)
```

```python
def issorted(arr):
    for i in range(1,len(arr)):
        if arr[i]<arr[i-1]:
            return False
    return True
arr=[1,2,3,45]
issorted(arr)
```

```python
def duplicate_arr(arr):
    i=0
    for j in range(1,len(arr)):
        if arr[j] != arr[i]:
            arr[i+1] = arr[j]
            i=i+1
    return arr[ :i+1]
arr=[1,1,2,3,4,4,5]
duplicate_arr(arr)
```

```python
def move_zero(arr):
    j=0
    for i in range(0,len(arr)):
        if arr[i] !=0:
            arr[j],arr[i] = arr[i],arr[j]
            j=j+1
    return arr
arr=[1,0,1,2,3,0,4,4,5]
move_zero(arr)
```

```python
def maxone(arr):
    count=0
    maxcount=0
    for i in range(len(arr)):
        if arr[i]==1:
            count=count+1
            maxcount=max(count,maxcount)
        else:
            count=0
    return maxcount
arr=[1,1,2,1,1,1]
maxone(arr)
```

```python
def single(arr):
    xor=0
    for i in range(len(arr)):
        xor=xor^arr[i]
    return xor    
arr=[1,2,1,2,3]
single(arr)
```

```python
def missing(arr):
    n=len(arr)
    hash_table=[0]*(n+1)
    for i in range(n):
        hash_table[arr[i]]=1
    for i in range(n+1):
        if hash_table[i] ==0:
            return i
arr=[0,2,3]
missing(arr)
```

```python
def missing(arr):
    n=len(arr)
    total=n*(n+1)//2
    partial=0
    for i in range(n):
        partial=partial+arr[i]
    return total-partial    
arr=[0,2,3]
missing(arr)
```

```python
def move_one(arr):
    temp=arr[0]
    for i in range(1,len(arr)):
        arr[i-1]=arr[i]
    arr[len(arr)-1]=temp
    return arr
arr=[1,2,3,4]
move_one(arr)
        
```

```python
def remove_val(arr,val):
    i=0
    for j in range(len(arr)):
        if arr[j] != val:
            arr[i]=arr[j]
            i=i+1
    return arr[ :i]
arr=[1,2,3,4,1,2,3]
val=3
print(remove_val(arr,val))     
```
