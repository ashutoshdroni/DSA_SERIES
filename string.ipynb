```python
#replace character in string
def replace (str,char1,char2):
    newstr=" "
    for char in str:
        if (char==char1):
            newstr = newstr+char2
        else:
            newstr = newstr+char
    return newstr
str = "abcrndjsbsdjsds"
str = replace(str,"s","d")
print(str)
```

```python
#cont vowels, constant, digit ,special character
def countinstring(str):
    v,c,d,s=0,0,0,0
    for char in str:
        if ((char>='a'and char<='z') or (char>='A' and char<="z")):
            char = char.lower()
            if (char=='a' or char=='e' or char=='i' or char=='o' or char=='u'):
                v=v+1
            else:
                c=c+1
        elif char >= '0' and char <= '9':
            d=d+1
        else:
           s=s+1
    return v,c,d,s
str = "/howAreyouashutosh124"
v,c,d,s=countinstring(str)
print(v,c,d,s)
        
    
```

```python
def ispalindrome(s):
    if len(s) <=1:
        return True
    if s[0]==s[-1]:
        return ispalindrome(s[1:-2])
    return False    
print(ispalindrome("sasa"))
print(ispalindrome("assa"))
```
