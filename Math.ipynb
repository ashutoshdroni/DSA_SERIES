```python
#extraction of digit
n=int(input())
while n>0:
    digit=n%10
    print(digit,end=" ")
    n=n//10
    
```

```python
#count digit in a number
n=int(input())
n=abs(n)    
count=0
while n>0:
    n=n//10
    count+=1
print(count)   
    
```

```python
#reverse a number
n=int(input())
n=abs(n)
rev=0
while n>0:
    digit=n%10
    rev=(rev*10)+digit
    n=n//10
print(rev)    
    
```

```python
#check a number is palindrome or not 
n=int(input())

dup=n
rev=0
while n>0:
    digit=n%10
    rev=(rev*10)+digit
    n=n//10
print(rev ==dup)
```

```python
#gcd(hcf)
def gcd(a,b):
    minimum=min(a,b)
    for i in range (minimum,0,-1):
        if a % i == 0 and b % i ==0:
            return i
print(gcd(10,20))            
```

```python
#armstong number
n=int(input())
dup=n
sum = 0
while n>0:
    digit=n%10
    sum=sum+(digit*digit*digit)
    n=n//10
print(sum ==dup)
```
