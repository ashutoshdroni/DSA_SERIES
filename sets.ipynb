```python
#intersection() - return common element in both set
a={1,2,3,4}
b={3,4,5,6}
print(a.intersection(b))   
```

```python
#union() - return elements from both sets without duplicate 
print(a.union(b))
```

```python
#difference() - return elements from first set but not in second 
print(a.difference(b))
```

```python
print(b.difference(a))
```

```python
#symmetric_difference() - return elements present in either set,but not in both
print(a.symmetric_difference(b))
```

```python
#sum of all unique number
def sumunique(l):
    s = set()
    for i in l:
        s.add(i)
    sum =0
    for i in s:
        sum = sum + i
    return sum
print(sumunique([1,2,3,4,4,5,5,1,2,3,4,5]))
```
