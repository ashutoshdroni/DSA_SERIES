```python
def num_1_to_n(n):
    for i in range(1,n+1,1):
        print(i)
num_1_to_n(5)        
```

```python
def num_n_to_1(n):
    for i in range(n,0,-1):
        print(i)
num_n_to_1(5)        
```

```python
def num_1_to_n(n):
    i=1
    while i<=n:
        print(i)
        i=i+1
num_1_to_n(5)         
```

```python
def fact(n):
    a= 1
    for i in range (1,n+1):
        a = a * i
    return a
print(fact(5))    
```

```python
def isprime(n):
    for i in range (2,n):
        if n%i == 0:
            break
    else:
         return True
    return False   
    
print(isprime(19))
```
